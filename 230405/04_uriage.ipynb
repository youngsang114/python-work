```python
#2. purchase_date 칼럼의 데이터를 시계열 데이터로 변경
# pd.to_datetime(변경할 데이터, format = "")
t = '2023-04-05'
pd.to_datetime(t, format="%Y-%m-%d")
```

```python
# 3. item_name 컬럼의 데이터를 모두 대문자로 변경
# upper() : 문자열을 모두 대문자로 변환
# str.upper() 사용하여 문자열을 대문자로 변경하는 방식


uriage['item_name'] = uriage['item_name'].str.upper()
```

```python
# apply() 함수를 이용하여 글자를 모두 소문자로 변경
# apply()는 스리즈형태데이터.apply(함수) ---> 스리즈의 데이터들을 하나씩 함수안에 넣어서 결과값으로 새로운 스리즈형태의 데이터로 생성한다
# apply()안의 함수는 만들어야됨
def change(x):
    result = x.lower()
    return result

uriage['item_name'] = uriage['item_name'].apply(change)
```

```python
# apply(lambda 함수) --->series, dataframe 모두에서 사용할 수 있다
# lambda 함수: 함수를 호출과 동시에 반환해주는 함수

uriage['item_name'] = uriage['item_name'].apply(lambda x : x.upper())   # x에 item_name 스리즈의 각각의 object 변수를 upper()함수로 실행한 후 반환해줌
```

```python
# 4. item_name에 있는 공백을 제거 한다.
# 공백을 제거 하는 함수 : strip() -> 좌,우 모두 공백 제거 ->현재 데이터에서 strip()은 사용 불가.
# 문자열 사이의 공백까지 제거하려면 사용할 함수의 이름?
# replace(기존 문자열, 대체할 문자열)
# replace(" ","")

uriage["item_name"].str.replace(" ","")
```

1. 상품 A의 item_price의 값은 결측치 중 상품이 A이면 이 값을 치환
2. 나머지 상품들도 같은 작업을 반복

```python
# 상품 A의 item_price을 출력
# 행의 조건? 결측치가 아니다, item_name이 상품A이다
# 열의 조건? item_price

# 결측치가 아닌 조건 
flg_null = uriage['item_price'].isna()  
flg_null     
#flg_null =uriage.isna().any(1)            # any()는 함수안에 우리가 구하고 싶은 값이 있는지 , # insa(): 결측치 ,(1): 행에 있는지 (0): 열에 있는지 

```

```python
# 결측치인 값들을 출력

uriage.loc[flg_null]
```

```python
# item_name이 상품 A인 경우만 출력
uriage.loc[uriage['item_name']=='상품A']
```

```python
# 상품이 A이고 item_price가 결측치인 경우
uriage.loc[(flg_null) & (uriage['item_name']=='상품A')]
```

```python
# 상품A이고 item_price가 결측치가 아닌 경우

price= uriage.loc[(uriage['item_name']=='상품A')& ~(flg_null),'item_price'].value_counts().index[0]
price
```

```python
# 결측치에 값을 치환하는 함수
# fillna(n): n의 값으로 결측치를 
uriage.loc[(flg_null) & (uriage['item_name']=='상품A')].fillna(price)
```

```python
uriage.loc[(flg_null) & (uriage['item_name']=='상품A'),'item_price'] = price
uriage
```

```python
#item_name의 리스트를 출력
name_list = uriage['item_name'].unique()
```

```python
for i in (uriage['item_name'].unique()):
    # item_pice의 값을 구한다.
    # item_name이 i와 같고 결측치가 아닌 데이터에서 price를 출력하는 과정
    price = uriage.loc[~(flg_null) & (uriage['item_name']==i),'item_price'].value_counts().index[0]

    # item_name이 i와 같고 결측치인 데이터에서 item_price에 price라는 변수를 삽입
    uriage.loc[(flg_null) & (uriage['item_name']==i) , 'item_price'] = price

```
