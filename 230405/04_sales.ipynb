```python
# series 형태의 데이터를 계산하고 , 새로운 파생변수에 저장
Sa_R['총단가']= Sa_R['판매단위'] * Sa_R['단가']
Sa_R['총원가']= Sa_R['판매단위'] * Sa_R['원가']
Sa_R['총이윤']= Sa_R['총단가'] - Sa_R['총원가']
```

```python
# 국가 커럼의 데이터가 항목들의 카운트를 출력하는 함수

Sa_R['국가'].value_counts()
```

```python
# 유니크한 국가 값
# 위의 series의 인덱스값 뽑기

Sa_R['국가'].value_counts().index
```

```python
len(Sa_R['국가'].unique())     #185개
_list= Sa_R['국가'].unique()          #위의 인덱스값 뽑는 것과 동일
```

```python
# 국가 하나를 선택해서 그 국가의 총이윤의 합계

Sa_R.loc[Sa_R['국가']=='South Korea']['총이윤'].sum()
```

```python
profit_list={}
for i in _list:
    # 총이윤의 합계
    total_profit = Sa_R.loc[Sa_R['국가']==i]['총이윤'].sum()
    profit_list[i]= total_profit
    

```

```python
profit_list
```

```python
# 딕셔너리형은 sort,sum등등의 함수가 없다
# 그래서 Series형태로 바꾼다음에 내림 차순으로 정렬하여서 최대값을 찾는다

pd.Series(profit_list).sort_values(ascending=False)
```

```python
# dictionary형 데이터를 DataFrame형태로!!!
# values, keys는 dictionary에서, dictionary의 없는 colums은 추가 지정

pd.DataFrame(profit_list.values(),index=profit_list.keys(),columns=['sum'])
```
