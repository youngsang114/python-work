```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
# drinks.csv 파일 로드
drinks= pd.read_csv("../csv/drinks.csv")
drinks.head()
```

```python
# 상관계수 행렬 출력

corr = drinks[
    [
    'beer_servings',
    'spirit_servings',
    'wine_servings',
    'total_litres_of_pure_alcohol'
    ]
].corr(method='pearson')
corr
```

```python
# 히트맵 출력

cols_view =['beer','spirit','wine','alchool']

sns.set(font_scale=1.5)

hm= sns.heatmap(
    corr.values,
    cbar= True,               # 색상에 대한 정보값,side bar
    annot=True,               # 히트맵 숫자 표시
    square=True,              # 정사각형 여부
    fmt='.2f',                # 소수점 표시   
    annot_kws={'size':20},    # 글자의 크기
    xticklabels= cols_view,   # x축의 이름
    yticklabels= cols_view    # y축의 이름

)

plt.tight_layout()
plt.show()
```

```python
sns.set(style='whitegrid', context= 'notebook')

sns.pairplot(
    drinks[
    [
    'beer_servings',
    'spirit_servings',
    'wine_servings',
    'total_litres_of_pure_alcohol'
    ]
],
height=2.5    
)

plt.show()
```
