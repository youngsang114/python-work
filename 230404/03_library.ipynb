```python
# 라이브러리 로드
import pandas as pd
```

```python
# 스리즈 형태 데이터를 생성
# 클래스로 만들어짐 Series
# 앞에 있는 값: index
# 뒤에 있는 값: values

pd_list= pd.Series([5000,6000,6500,6500],                             
                   index=['아메리카노','카페라떼','카페모카','카푸치노']) 
```

```python
print(pd_list)  
#dtype: value의 형태가 int
```

```python
print('스리즈 형태의 데이터에서 index 의 값은 :', pd_list.index)
print('스리즈 형태의 데이터에서 value 의 값은 :', pd_list.values)
```

```python
# 2차원 데이터셋
# 데이터 프레임 생성
# DataFrame은 Series들의 합이다.

_value=[[1,2,3],[4,5,6],[7,8,9]]
_index=['a','b','c']
_column=['A','B','C']

df= pd.DataFrame(_value,index=_index,columns=_column)
# print(df)
df #데이터를 보여주는 것
```

```python
print("데이터 프레임의 인덱스의 값 :" ,df.index)
print("데이터 프레임의 컬럼의 값 :", df.columns)
print("데이터 프레임의 벨류의 값 :", df.values)
```

# 넘파이(Numpy)

- 수치 데이터를 다루는 라이브러리, 선형 대수 계산에서 주로 사용된다.
- 함수로 불러온다, 편의성 속도면에서 우수하다 

#### array()
#### numpy로 만드는 데이터는 ndarray형태이다

- 리스트 튜플 배열로부터 ndarray를 생성한다 
- 1차원 2차원 배열이든 상관없이 array를 생성할 수 있다.
- 축의 개수(ndim), 크기(shape)를 출력할 수 있다 --->튜플 형태로 값을 반환한다.
- zeros()는 모든 원소의 값이 0인 배열을 생성한다.
- ones()는 모든 원소 값이 1인 배열을 생성한다.
- full((크,기),어떤 값)으로 행렬을 생성한다.
- eyes()는 단위 행렬을 생성한다.
- random.random()은 임의의 값을 원소로 하여 행렬을 생성한다.
- arrange(n)은 0부터 n-1까지 배열을 만들어 준다. arrange(i,j,k) !!!
- 사칙연산
    1. np.add(더하기)
    2. np.subtract()
    3. np.multiply()
    4. np.divide()
- 행렬곱
    1. dot()이라는 함수 사용
    2. @가 행렬곱이다.
    3. (a,b) @ (b,c) = (a,c)
    


```python
# numpy는 array([])형태로 바꿔서 사용해야한다.

import numpy as np

np_array_1= np.array([1,2,3,4,5])
print(np_array_1)
```

# 맷플롯립(Matplotlib)

- 파이썬에서 데이터를 차트,플롯으로 그려주는 라이브러리, 데이터 시각화 패키지

#### .plot()

- 하얀색 도화지에 수치 레이어를 올리고, 옵션을 레이어 하는 함수
- plot('x축 리스트','y축 리스트','옵션')  
- 색(b,g,r,c,m,y,k,w)
- 선 종류('-','--','o','v','-.','.',',','*','x','D')
- 레이블,타이틀 추가 ----> plt.xlable() , plt.ylable(), plt.title()
- 그래프 라인의 범례 추가 --> plt.legend()
- 범례에 지정된 위치에 추가하기(loc="위치") ---> plt.legend([범,례],loc="upper right" )
    - 위치는 string, code 값으로 지정 가능하다 (기본값="best",0)
- 서브플롯 subplot---> 하나의 도화지에 여러개의 그래프를 나눠줌
    - plt.subplot(nrow,ncol,pos)
    - nrow: 행의 수
    - ncol: 열의 수
    - pos: 위치


#### .bar()

- 바그래프
- x축 필수, x축 y축 개수가 일치해야 됨
- xticks()---> x축의 이름(레이블)을 변경 가능
- 옵션 
    - 색상 color="" 리스트 형태로 각각 색상을 만들어 줄 수 있다.
    - 폭 지정 widht='0.8'

#### scatter()

- 산점도 
- plt.scatter(x,y,옵션)
- area,colors
