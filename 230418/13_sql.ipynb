```python
import mod_sql
import pandas as pd
```

```python
# mod_sql을 통해서 Database 클래스를 생성
db=mod_sql.Database("localhost","root","1234","ubion7",3306)
```

```python
sql="""
    select
    `Country` as '국가',
    `Units Sold` as '판매개수',
    `Unit Price` as `가격`,
    `Unit Cost` as `원가`,
    (`Units Sold` * `Unit Price`) as `총수익`,
    (`Units Sold` * `Unit Cost`) as '총비용',
    ((`Units Sold` * `Unit Price`) - (`Units Sold` * `Unit Cost`)) as `총이윤`
    from 
    `Sales Records`

"""

db.sql_query(sql)
```
