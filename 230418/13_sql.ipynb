1. emp 데이터프레임에서 JOB컬럼의 데이터를 확인하여 새로운 파생변수를 생성
2. job_code -->SALESMAN이면 1 MANAGER이면 2 나머지는 3으로 하여 새로운 파생변수에 대입

```python
#emp["JOB"]
```

```python
emp=pd.read_csv("../csv/emp.csv")
emp['JOB_CODE']=''

for i in emp.index:
    if emp.loc[i,'JOB']=='SALESMAN':
        emp.loc[i,'JOB_CODE']=1

    elif emp.loc[i,'JOB']=='MANAGER':
        emp.loc[i,'JOB_CODE']=2
        
    else:
        emp.loc[i,'JOB_CODE']=3
    
```

```python
emp["JOB_CODE"]=emp['JOB'].apply(lambda x : 1 if (x=='SALESMAN') else (2 if (x=='MANAGER') else 3 ))
```

```python
def change(x):
    if x=='SALESMAN':
        result =1
    elif x=='MANAGER':
        result=2
    else:
        result=3
    return result

emp['JOB_CODE']=emp['JOB'].apply(change)
```

```python
sql="""
    select
    `EMPNO`,
    `ENAME`,
    `SAL`,
    if(
        `SAL`>=2000,
        "HIGH",
        "LOW"
    )as `H/L`
    from 
    emp
"""
db.sql_query(sql)
```

### 테이블의 결합
1. 유니언 결합(단순하게 행을 결합)
2. 조인 결합

```python
sql="""
    select
    *
    from
    tran_1

    union
    
    select
    *
    from
    tran_2
"""
db.sql_query(sql)
```

```python
sql="""
    select
    *
    from
    tran_d_1

    union

    select
    *
    from
    tran_d_2
"""
db.sql_query(sql)
```

```python
sql="""
    select
    *
    from
    tran_1
    left join
    tran_d_1
    on
    tran_1.transaction_id = tran_d_1.transaction_id
"""
db.sql_query(sql)
```

```python
sql="""
    select * from
    (select * from tran_1 
    union
    select * from tran_2) as tran
    left join
    (select * from tran_d_1
    union
    select * from tran_d_2) as tran_d
    on tran.transaction_id = tran_d.transaction_id
"""
db.sql_query(sql)
```

```python
tran_1=pd.read_csv("../csv/tran_1.csv")
tran_2=pd.read_csv("../csv/tran_2.csv")
tran_d_1=pd.read_csv("../csv/tran_d_1.csv")
tran_d_2=pd.read_csv("../csv/tran_d_2.csv")
```

```python
tran=pd.concat([tran_1,tran_2],axis=0)
tran_d=pd.concat([tran_d_1,tran_d_2],axis=0)
```

```python
pd.merge(tran,tran_d,left_on="transaction_id", right_on="transaction_id",how='left')

### 열의 이름이 다를경우 left_on, right_on을 써서 다른 이름의 열에 같은 value값을 left로 merge해준다
```
