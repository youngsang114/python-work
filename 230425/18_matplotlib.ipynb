# 데이터의 시각화
1. 라인, 바, 산점도, 히스토그램 그래프 시각화

```python
import matplotlib.pyplot as plt
```

```python
# 라인 그래프
# plt.plot([])-->리스트의 값이 하나만 존재하면, y축의 값으로 지정
# plt.plot([],[])--> 앞의 리스트의 값은 x축, 뒤의 리스트의 값은 y축이 된다.
# x축과 y축의 개수가 같아야 됨 -->아니면 오류남!

plt.plot([1,2,3,4])
plt.show()
```

## subplot()
1. 한 영역에 그래프를 여러 개 그려주기 위한 함수
2. subplot(행의 개수, 열의 개수, 위치 값)

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4])

plt.subplot(1,2,2)
plt.plot([4,3,2,1])

plt.show()
```

```python
fig,axes =plt.subplots(2,2)

axes[0][0].plot([1,2,3,4])
axes[0][1].plot([4,3,2,1])
axes[1][0].plot([1,2,2,1])
axes[1][1].plot([3,4,2,1])
```

```python
import platform
# 그래프에서 한글 깨짐 방지
if platform.system() == "Darwin":          # MAC인 경우
    plt.rc('font', family ="AppleGothic")
else:
    plt.rc('font',family='Malgun Gothic')
```

```python
# x축의 이름과 y축의 이름, 그래프의 이름 설정

plt.plot([1,2,3,4])
plt.xlabel('x축')
plt.ylabel('y축')
plt.title('title')

plt.show()
```

```python
# 범례 추가
plt.plot([1,2,3,4],label='acending')
plt.plot([4,3,2,1],label='descending')
plt.legend(loc='upper right')
plt.show()
```

```python
# 히스토 그램 -> 데이터의 개수를 표현하는 바 그래프
data=[1,2,2,2,3,3,4,4,4,4]
plt.hist(data,bins=4)
# bins는 막대 개수
plt.show()
```
