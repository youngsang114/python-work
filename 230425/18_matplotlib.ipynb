## subplot()
1. 한 영역에 그래프를 여러 개 그려주기 위한 함수
2. subplot(행의 개수, 열의 개수, 위치 값)

```python
plt.subplot(1,2,1)
plt.plot([1,2,3,4])

plt.subplot(1,2,2)
plt.plot([4,3,2,1])

plt.show()
```

```python
fig,axes =plt.subplots(2,2)

axes[0][0].plot([1,2,3,4])
axes[0][1].plot([4,3,2,1])
axes[1][0].plot([1,2,2,1])
axes[1][1].plot([3,4,2,1])
```

```python
from datetime import datetime
```

```python
mean_kospi_2020=kospi.loc['2020'].resample("MS").mean()
mean_kospi_2020
```

```python
# x축은 'xxxx년 xx월' 표시
mean_kospi_2020.index[0].strftime("%Y년%m월")

x=[]
for i in mean_kospi_2020.index:
    x.append(i.strftime("%Y월%m월"))

x
```

```python
# 인덱스 값을 문자열로 저장-->map함수 이용
list(map(lambda x : x.strftime("%Y년%m월"), mean_kospi_2020.index))
```

```python
plt.barh(x,mean_kospi_2020['Volume'])
# 지수 표현식을 일반 표현식으로 변경
plt.ticklabel_format(axis='x',style='plain')
plt.show()
```

```python
# kospi에서 변동률 파생변수 생성
# 변동률 =(종가-시가)/시가 * 100

kospi['변동률']=(kospi['Close']-kospi['Open'])/kospi['Open']* 100
kospi.head()
```

```python
# 음수를 표시
plt.rc('axes',unicode_minus=False)

plt.grid()
plt.hist(kospi['변동률'],bins=20)
plt.show()
```

# 문제 
1. test_csv폴더 안에 있는 오프라인.csv을 로드
2. 컬럼의 이름을 ['기준년월','업종','성별','연령대','총이용자수','총이용금액']으로 변경
3. 기준년월을 시계열 데이터로 변경
4. index를 기준년월로 변경
5. 온라인.csv파일도 같은 1~4 작업을 반복한다
6. 온라인에 업종 컬럼을 추가하여 '온라인'으로 값을 지정하고
7. 두개의 데이터프레임을 유니언 결합
8. 기준년월, 업종별로 총 이용금액의 합계를 그래프로 시각화

```python
import pandas as pd
```

```python
df=pd.read_csv("../test_csv/오프라인.csv",encoding="euc-kr")
df.columns=['기준년월','업종','성별','연령대','총이용자수','총이용금액']
df
```

```python
df['기준년월']=pd.to_datetime(df['기준년월'],format="%Y%m")
```

```python
df.set_index('기준년월',inplace=True)
```

```python
df
```

```python
df2=pd.read_csv("../test_csv/온라인.csv",encoding="euc-kr")
df2.columns
```

```python
df2['업종']='온라인'
df2
```

```python
df2.columns=['기준년월','성별','연령대','총이용자수','총이용금액','업종']
```

```python
df2['기준년월']=pd.to_datetime(df2['기준년월'],format="%Y%m")
```

```python
df2.set_index('기준년월',inplace=True)
```

```python
data=pd.concat([df,df2],axis=0)
data
```

```python
pivot_data=pd.pivot_table(data,
               index=data.index,
               columns='업종',
               values='총이용금액',
               aggfunc='sum')
pivot_data.plot()
plt.show()
```

```python
group_data2=data.groupby([data.index,'업종']).sum()[['총이용금액']]
```
