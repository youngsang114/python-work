## 할로윈 투자 전략
1. 11월 1일에 구매  -> 6개월간 유지 4월 말일에 판매
2. 11월 첫날 시가로 구매
3. 4월 마지막날의 종가로 판매
4. 수익률 =(3번과정-2번과정)/2번과정

```python
from datetime import datetime
```

```python
df=pd.read_csv("../csv/AMZN.csv")
df.set_index("Date",inplace=True)
df
```

```python
df.index=pd.to_datetime(df.index,format="%Y-%m-%d")
df["Month"]=list(map(lambda x: datetime.strftime(x, '%m'),df.index))
df['Month']=df['Month'].astype(int)
```

```python
New_df=df.loc[(df["Month"]==11) | (df["Month"]==4)]
New_df.head()
```

```python
New_df["Y-M"]=list(map(lambda x: datetime.strftime(x, '%Y-%m'),New_df.index))
data=New_df.loc[((New_df["Y-M"] != New_df.shift(-1)["Y-M"]) & (New_df['Month']==4)) | ((New_df["Y-M"] != New_df.shift(1)["Y-M"]) & (New_df['Month']==11))]
New_data=data.iloc[6:,]
New_data
```

```python
New_data['TRADE']=''

for i in New_data.index:

    if New_data.loc[i,'Month']==11:
        New_data.loc[i,'TRADE']='Buy'
    else:
        New_data.loc[i,'TRADE']='Sell'
New_data.head(5)
```

```python
rtn=1
buy=0
sell=0

New_data['rtn']=1

for i in New_data.index:
    if New_data.loc[i,"TRADE"]=='Buy':
        buy=New_data.loc[i,'Open']
        #print(buy)
    else:
        sell=New_data.loc[i,'Close']
        #print(sell)
        rtn=(sell-buy)/buy+1
        #print(rtn)
        New_data.loc[i,'rtn']=rtn

New_data
```

```python
acc_rtn=1.0
for i in New_data.index:
    acc_rtn*=New_data.loc[i,'rtn']
print(acc_rtn)
```

```python
acc_rtn=1
for i in range(2000,2019,1):
    buy_mon= str(i)+"-11"
    sell_mon=str(i+1)+"-04"
    
    # 구입가 저장
    buy=df.loc[buy_mon].iloc[0]['Open']
    sell=df.loc[sell_mon].iloc[-1]['Close']
    
    #수익율
    rtn=(sell-buy)/buy+1
    # 누적 수익율
    acc_rtn*=rtn
    
print(acc_rtn)
```

```python
# buy and hold인경우
buy=df.loc["2000-11"].iloc[0]['Open']
sell=df.loc["2019-04"].iloc[-1]["Close"]

rtn=(sell-buy)/buy+1
print(rtn)
```

```python
# CAGR(연복리 수익율)
# (수익율**(1/투자기간)) -1

# 할로윈 CAGR
CAGR=(acc_rtn **(1/20))-1
print(CAGR*100)
```

### 함수화


```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
start=datetime(year=2020,month=11,day=1)
start
```

```python
end=start+relativedelta(months=5)
end
```

```python
print(start.strftime("%Y-%m"),end.strftime('%Y-%m'))
```

```python
def  six_month(df,start=2000,end=2019,month=11):
    if 'Date' in df.columns:
        df['Date']=pd.to_datetime(df["Date"],format="%Y-%m-%d")
        df.set_index('Date',inplace=True)

    # 누적 수익율 변수를 생성
    acc_rtn=1

    for i in range(start,end):
        start_= datetime(year=i,month=month,day=1)
        end_= start_+relativedelta(months=5)

        buy_mon=start_.strftime("%Y-%m")
        sell_mon=end_.strftime("%Y-%m")

        buy=df.loc[buy_mon].iloc[0]["Open"]
        sell=df.loc[sell_mon].iloc[-1]["Close"]

        rtn=(sell-buy)/buy + 1
        
        acc_rtn*= rtn
    return(acc_rtn)
```

```python
df_=pd.read_csv("../csv/AMZN.csv")
```

```python
six_month(df_)
```
