```python
import requests
from bs4 import BeautifulSoup as bs
import pprint as pp
```

```python
# 실제 사이트에서 html 정보를 받아와서 변수에 저장
data=requests.get("https://finance.naver.com/marketindex/exchangeList.naver").text
```

```python
# BeautifulSoip을 이용하여 데이터를 파싱
soup = bs(data, 'html.parser')
```

```python
# find는 html안에서 해당 태그의 첫 번째 정보를 출력

tbody = soup.find('tbody')
```

```python
# find_all() 해당 태그의 전체 정보를 리스트의 형태로 출력
data_list=soup.find_all('td')
data_list
```

```python
# 1차원 리스트 생성
_list =[]
for i in data_list:
    x = i.get_text()
    x = x.replace('\n','')
    x = x.strip('\t')
    _list.append(x)
_list
```

```python
result=[]
for i in range(0,(len(_list)//7),1):
    x=[]
    for j in range(i*7,(i+1)*7,1):
        x.append(_list[j])
    result.append(x)

pp.pprint(result)
```

```python
result=[]
for i in range(len(_list)//7):
    result.append(_list[i*7:(i+1)*7])
result
```
