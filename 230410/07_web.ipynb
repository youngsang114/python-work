### 웹 서버란?
1. 하드웨어: 실제 물건을 만드는 주방, 웹 사이트의 컴포넌트 파일들을 저장하는 컴퓨터
2. 소프트 웨어: 음식점의 레시피, 어떻게 호스트 파일들에 접근하는지 관리

- 웹 서버에서 불려진 파일을 필요로 할때, 브라우저는 HTTP로 파일을 요청, 요청이 웹 서버에 도달할 때, HTTP 서버는 요청된 문서를 HTTP를 이용해 보낸다
    - 정적 웹 페이지
        - 고정된 페이지, 웹 서버에 미리 저장된 파일이 그대로 전달되는 웹 페이지
        - 속도가 빠르고 비용이 적게듬
        - 서비스가 한정적이다, 페이지 수정,추가,삭제 모두 수동 -->유지 보수 힘들다
    - 동적 웹 페이지
        - 웹 서버에 있는 데이터들을 스크립트에 의해 가공 처리 후 생성되어 전달되는 웹 페이지
        - 유저는 상황,시간,요청에 따라 달라지는 웹페이지를 보게 됨
        - 서비스가 다양하다,페이지 수정,추가,삭제 모두 쉽다--> 관리가 쉽다
        - 속도가 상대적으로 느리다, 웹 서버 외 어플리케이션 서바(데이터 베이스 서버)가 필요하므로 추가 비용이 발생한다
