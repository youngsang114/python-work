### 수익율 계산
1. 파생변수(return) 값은 1로 생성
2. rtn =1.0 ,sell변수는 0으로 생성
3. 반복문을 이용해서 현재의 trade가 buy고 전 행의 trade가 ''인 경우 구매 날
4. 전 행의  trade가 buy이고 현재의 trade가 ''인 경우 판매 날
5. 수익율 계산하여 return에 대입

### 누적 수익율 계산
1. acc_rtn=1.0 생성
2. return의 항목들을 누적 곱하여 acc_rtn컬럼에 대입
3. 최종적으로 acc_rtn을 출력

```python
df['return']=1.0

rtn=1.0
buy=0.0
sell=0

# index의 데이터 값이 unique해서 
for i in df.index:         
    # 구매한 날을 체크, 구매가 지정
    if (df.loc[i,"trade"] == "buy") and (df.shift(1).loc[i,"trade"]== ""):
        buy=df.loc[i,"Adj Close"]
        # print (buy)
        # 이 시점에 사고 그 값을 변수 buy 값에 저장해놓는다
        # print("구매일:",i ,'구매가격',buy)

    # 판매한 날을 체크, 판매가 지정, 수익률 계산
    elif (df.shift(1).loc[i,"trade"]=="buy") and (df.loc[i,"trade"]==""):
        sell= df.loc[i,"Adj Close"]
        rtn= (sell-buy)/buy +1
        df.loc[i,'return']=rtn
        # print("판매일:",i ,'판매가격',sell)

        # 이 시점에 팔거야 근데 산 시점은 위의 if문일때 샀음--> 수익률을 팔 시점의 판 값(이 건 위 for문에 저장되어 있음), 산 시점의 산 값을 변수에 저장해놓는다
        # 그렇게 나온 값을 rtn에 저장해놓고, 판 시점의 df['rtn']에 대입한다
        # 다시 사는 것부터 반복

acc_rtn=1.0

for i in df.index:
    acc_rtn *= df.loc[i,'return']
    df.loc[i,'acc_rtn']= acc_rtn
    
print('누적 수익율:',acc_rtn)
```

# 절대 모멘텀을 함수화
1. 첫 번째 함수
    - 매개변수 1개(데이터 프레임)
    - 데이터프레임에 결측치와 이상치를 제거
    - 수정종가를 제외한 나머지 컬럼은 삭제
    - 인덱스를 시계열로 변경하기
    - "STD-YM" 파생변수를 생성하여 인덱스의 년-월 추출해서 대입
2. 두 번째 함수
    - 매개변수 1개(데이터 프레임)
    - 새로운 데이터프레임(df2) 생성
    - 인자값으로 받아온 데이터프레임에서 년-월별 마지막 데이터들을 새로운 데이터프레임에 대입
    - df2에 파생변수 2개 생성
        - 'BF_1M': 전월의 종가, 결측치는 0으로 대입
        - "BF_12M": 전 년도의 종가, 결측치는 0으로 대입
        - return df2
3. 세 번째 함수
    - 매개변수 2개(첫번째 함수의 결과(df1), 두번째 함수의 결과(df2))
    - df1에 trade 컬럼을 생성, 값은 ''
    - df1에 return 칼럼을 생성, 값은 1
    - df2의 값들을 이용하여 momnentum_index를 구하고 df1에 거래 내역 삽입
    - df1의 거래 내역을 이용하여 수익율 return 컬럼에 대입
    - return 컬럼의 데이터를 가지고 누적수익율(acc_rtn)에 대입
    - 누적 수익율을 print를 이용하여 출력
    - df1을 리턴

#### 함수화 코드

```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv("../csv/GDX.csv",index_col='Date')
df
```

```python
def momentum1(df,col):
    
    # 만약에 인덱스가 Date가 아니면? Date컬럼을 인덱스로 변경
    if 'Date' in df.columns:
        df=df.set_index("Date")

    # 결측치와 이상치를 제거
    df=df.loc[~df.isin([np.nan,np.inf,-np.inf]).any(axis=1)]
    # 수정종가를 제외한 나머지 컬럼 삭제
    df=df[[col]]
    # 인덱스를 시계열로 변경
    df.index=pd.to_datetime(df.index,format="%Y-%m-%d")
    # 'STD-YM'컬럼을 생성
    df['STD-YM']=list(map(lambda x: datetime.strftime(x,'%Y-%m'),df.index))

    return df

```

```python
# 첫번째 함수 실행해보기
momentum1(df,'Adj Close')
```

```python
def momentum2(df):

    df2= pd.DataFrame()
    # 리스트값을 만들어서 for문을 돌림
    _list=df["STD-YM"].unique()
    col=df.columns[0]

    

    for i in _list:
        last_df=df.loc[df["STD-YM"] ==i].tail(1)
        df2 =pd.concat([df2, last_df])

    df2['BF_1M']=df2.shift(1)[col].fillna(0)
    df2['BF_12M']=df2.shift(12)[col].fillna(0)
    
    return df2

```

```python
def momentum3(df1,df2):

    col=df1.columns[0]
    df1['trade']=''
    
    for i in df2.index:
        signal=''

        # 절대 모멘텀 계산
        monentum_index= (df2.loc[i,"BF_1M"]/ df2.loc[i,"BF_12M"]) - 1
        
        # 절대 모멭넘 지표에 따라서 True / Fasle 로 구분
        flag = True if((monentum_index >0 ) and (monentum_index!=np.inf) and (monentum_index !=-np.inf)) else False

        if flag:
            signal ='buy'
        #print('날짜:',i, "모멘텀 인덱스 :", monentum_index, 'flag :',flag, 'signal :', signal)

        df1.loc[i,'trade']=signal

        df1['return']=1.0

    rtn=1.0
    buy=0.0
    sell=0

    # index의 데이터 값이 unique해서 
    for i in df1.index:         
        # 구매한 날을 체크, 구매가 지정
        if (df1.loc[i,"trade"] == "buy") and (df1.shift(1).loc[i,"trade"]== ""):
            buy=df1.loc[i,col]
            # print (buy)
            # 이 시점에 사고 그 값을 변수 buy 값에 저장해놓는다
            # print("구매일:",i ,'구매가격',buy)

        # 판매한 날을 체크, 판매가 지정, 수익률 계산
        elif (df1.shift(1).loc[i,"trade"]=="buy") and (df1.loc[i,"trade"]==""):
            sell= df1.loc[i,col]
            rtn= (sell-buy)/buy +1
            df1.loc[i,'return']=rtn
    acc_rtn=1.0

    for i in df1.index:

        acc_rtn *= df1.loc[i,'return']
        df1.loc[i,'acc_rtn']= acc_rtn
        
    print('누적 수익율:',acc_rtn)
    return df1
        
```

```python
df=momentum1(df,'Adj Close')
df2=momentum2(df)
momentum3(df,df2)
```
